```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
import random
```

```python
data = [random.choice([0, 1]) for _ in range(100)]
grouped_data = [np.mean(data[i:i+10]) for i in range(0, len(data), 10)]  
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
plt.plot(grouped_data, color='orange')
plt.xlabel("Time")
plt.ylabel("Posture")
plt.show()
```
