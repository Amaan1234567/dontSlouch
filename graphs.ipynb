```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
import random
```

```python
data = [random.choice([0, 1]) for _ in range(10_000)]
```

```python
grouped_data = [np.mean(data[i:i+100])*100 for i in range(0, len(data), 100)]
```

```python
plt.plot(grouped_data)
```

```python
fig, ax = plt.subplots()
line1 = ax.plot(0, grouped_data[0])[0]
ax.set(xlim = [0, 10000], ylim = [0, 100])
def update(frame):
    x = [i*1000 for i in range(frame)]
    y = grouped_data[:frame]
    line1.set_xdata(x)
    line1.set_ydata(y)
    return line1
ani = animation.FuncAnimation(fig, update, frames=100, interval=1)
ani.save("test.gif", writer = "pillow")
```
